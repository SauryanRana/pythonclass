```python
from bs4 import BeautifulSoup
import requests
url= BeautifulSoup("https://www.worldometers.info/coronavirus/",'html.parser')
request= requests.get(url)
request
```

```python
data = request.text #(print the html text froml website)
lxml_data= BeautifulSoup(data,"lxml")
lxml_data
```

```python
raw data= lxml_data.table.text
```

```python
raw_data= lxml_data.table
tags = raw_data.find_all('tr')
tags
```

```python
final_data= []

for i in tags:
    final_data.append(i.text.split('\n')[1:-1])
    
print(final_data)
```
